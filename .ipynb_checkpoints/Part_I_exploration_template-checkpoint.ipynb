### Cleaning Data For Analysis

```python
#copy of dataframe
bike = data.copy()
```

> I created a copy of the dataframe because I did not want to alter the original data. I will convert the datatype of member_birth_year to integer as it will be used for calculation to generate an age feature, then I will extract the date from the start_time column, after which I will select only columns needed for analysis

```python
#convert type from object to integer and Nan to 0
bike['member_birth_year'] = bike['member_birth_year'].fillna(0).astype(int)
#generate age column
bike["age"] = bike["member_birth_year"].apply(lambda x: 2019 - x if x > 0 else 0)
```

```python
#extracting date of the month
bike['date'] = bike['start_time'].map(lambda x: x[:10])
bike['date'] = bike['date'].map(lambda x: datetime.strptime(x, '%Y-%m-%d'))
#selecting date from datetime
bike['date'] = bike['date'].map(lambda x: x.day)
```

> Next I will drop rows that have Nans and then select only columns I will be using for the analysis

```python
# drop Nans
biketrip = bike.dropna()
```

```python
biketrip = biketrip[['duration_sec', 'bike_id', 'user_type', 'age',  'member_gender', 'date', 'bike_share_for_all_trip']]
```

```python
#shape of dataframe
biketrip.head()
```

## Univariate Exploration

> In this section, I will investigate the distributions of individual variables to see how each feature is distributed. I will start with the main feature of interest duration_sec


#### Duration_sec
> How is duration_sec distributed? I will start by plotting a histogram to show distribution of duration_sec

```python
plt.hist(data=biketrip, x='duration_sec')
plt.xlabel('Duration_sec (seconds)')
plt.ylabel('Count')
plt.title('Duration_sec Distribution');
```

> from the histogram above, I can not gather any intelligent insight, so I will change the x axis to logarithmic scale to see if there will be change

```python
cut = 10 ** np.arange(1.7, 4.3+0.1, 0.1)
plt.hist(data=biketrip, x='duration_sec', bins=cut)
plt.xscale('log')
plt.xticks([100, 250, 500, 1000, 2500, 5000, 10000],[100,250, 500, 1000, 2500,5000, 10000] )
plt.xlabel('Duration_sec (seconds)')
plt.ylabel('Count')
plt.title('Duration_sec Distribution Logarithmic Scale');
```

>After transfroming the x axis scale into logarithmic scale, the histogram is now understandable as I can gather knowlegdeable insights from it. The Duration_sec feature is unimodal, having one peak around 500-600 seconds and it is also fairly symmetrical. I am really curious on factors that contribute to duration_sec distibution. I can also see there are outliers, so I will replace the extreme values with the mean of duration_sec

```python
#Cleaning outliers
Q25 = biketrip['duration_sec'].quantile(0.25)
Q75= biketrip['duration_sec'].quantile(0.75)
diff = Q75 - Q25
# detect extereme values
Q1 = (biketrip['duration_sec'] > (Q75 + 1.5 * diff))
extereme = biketrip['duration_sec'][~Q1].max()
#replace with mean 
biketrip['duration_sec'] = np.where(biketrip['duration_sec']> extereme, biketrip['duration_sec'].mean() , biketrip['duration_sec'])
```

#### Member_gender
> How is the member_gender feature distributed?

```python
# function to plot graph
def Visualize(feature, title):
    """ Visualize distribution of the feature"""
    #select a color to display
    color = sb.color_palette()[0]
    #ordering bars
    desc = biketrip[feature].value_counts().index
    #barchart
    sb.countplot(data=biketrip, x=feature, color=color, order=desc)
    plt.title(title);
    #calculate percentage
    total = biketrip[feature].value_counts().sum()
    unit = biketrip[feature].value_counts()
    #generate label and ticks
    locs,labels = plt.xticks()
    ticks = zip(locs,labels)
    # to display percentage of the different categories
    for loc, label in ticks:
        x = unit[label.get_text()]
        percent = '{:0.1f}%'.format(100*x/total)
        plt.text(loc,x+2,percent, ha='center',color='black')
```

```python
#generate bar chart for member_gender
Visualize('member_gender', 'Distribution of member_gender')
```

> From the visualization, I can clearly see that the male gender is predominant having a very large share of the rider's population, It is even more than the female and other gender combined. The percentage of Male gender is 74.6%, while that of Female is 23.3%. The Other gender comes last with 2.1%.  Are the male riders responsible for the duration_sec peak? Do they take longer rides? I will find later in the analysis

## Multivariate Exploration

> I will create a graph showing the relationship between three different variables in the multivariate exploration

#### Duration_sec, Bike_share and User_type
> How are these three features related? Do riders take longer rides when bike is shared? Which category of User_type share bike?

```python
#generate pointplot
plt.figure(figsize=[15, 5])
ax = sb.pointplot(data = biketrip, x = 'user_type', y = 'duration_sec', hue = 'bike_share_for_all_trip',
                  dodge = 0.1, linestyles = "")
plt.title('Duration_sec, Bike_share and User_type');
```

> From the pointplot above, only the subscriber category of user_type share rides and rides where bike were shared had the lowest duration_sec. So sharing bike does not increase the duration of a trip
